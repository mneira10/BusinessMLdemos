```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
train = pd.read_csv('train.csv')
```

```python
train
```

```python
subTrain = train[train.store==1]
```

```python
fig = go.Figure()

for item in subTrain.item.unique():
    temp = subTrain[subTrain.item==item]
    plt.scatter(temp.date,temp.sales)
    fig.add_trace(go.Scatter(x=temp.date, y=temp.sales,
                    mode='lines',
                    name='Product {}'.format(item)))
    break

fig.show()
```
